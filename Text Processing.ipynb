```python
#reading the csv file
import nltk
import pandas as pd
data=pd.read_csv("twitter_trends.csv")
data.head()
```

```python
#converting it to a string
nltk.download()
trends=list(data['Indian trends'])
s=''
for each in trends:
    s=s+each+' ';
print(s)    
```

```python
#word tokenization 

from nltk.tokenize import word_tokenize
tokenized_list=word_tokenize(s)
print(tokenized_list)
```

```python
#stemming
df = pd.DataFrame()
df['OriginalWords'] = pd.Series(tokenized_list)
#porter's stemmer
porterStemmedWords = [nltk.stem.PorterStemmer().stem(word) for word in tokenized_list]
df['PorterStemmedWords'] = pd.Series(porterStemmedWords)
#SnowBall stemmer
snowballStemmedWords = [nltk.stem.SnowballStemmer("english").stem(word) for word in tokenized_list]
df['SnowballStemmedWords'] = pd.Series(snowballStemmedWords)
# LancasterStemmer
lancaterStemmerWords=[nltk.stem.LancasterStemmer().stem(word) for word in tokenized_list]
df['LancasterStemmedWords'] = pd.Series(lancaterStemmerWords)
df
```

```python
#n-grams upto 3 grams
from nltk.util import ngrams
Bigrams=ngrams(tokenized_list,2)
list(Bigrams)
```

```python
trigrams=ngrams(tokenized_list,3)
list(trigrams)
```

```python
#document term matrix on stemmed words
dtm={}
for each in porterStemmedWords:
    if each in dtm.keys():
        dtm[each]=dtm[each]+1
    else:
        dtm[each]=1
document_term_matrix = pd.DataFrame()
document_term_matrix['word'] = pd.Series(list(dtm.keys()))
document_term_matrix['frequency'] = pd.Series(list(dtm.values()))
document_term_matrix
```
